### Renta per capita

Tenemos un dataset que contiene la renta per cápita disponible bruta de Madrid, tanto en general como por distritos, para los años comprendidos entre 2002 y 2019.

Primero, realizamos una limpieza de la tabla para conservar únicamente las fechas, los nombres de las columnas y los distritos, manteniendo el formato que se encuentra en otros datasets.

A continuación, pivotamos la tabla de manera que los distritos se conviertan en columnas y descargamos el archivo resultante.

Luego, llevamos a cabo un análisis para prever cómo será la renta en los años siguientes y comprobar si sigue un patrón lógico con el que podamos trabajar.

Después, implementamos un bucle para realizar la regresión en cada columna y para los tres años necesarios.

Descargamos nuevamente el dataset con los resultados obtenidos.

A continuación, aplicamos la operación "melt" para transformar el dataset en uno más largo y lo descargamos nuevamente.


```python
# Librerias 

import pandas as pd
import numpy as np
```

```python
# Leer el excel de la tabla extraida de otro dataset mas grande y detallado por que solo necesitamos los datos de la renta per capita por distrito
df = pd.read_excel('/Users/Paco/IMMUNE/Modulo_Deep_Learning_Completo 2/Modulo_Deep_Learning_Completo/notebooks/Renta_per_capita.xlsx', header=1) # La primera fila es el nombre de las columnas que son los años
df
```

```python
# La primera columna le ponemos el nombre distrito
df = df.rename(columns={'Unnamed: 0': 'distrito'})
```

```python
df
```

```python
# cambiar los nombres de los distritos a minuscula y quitar la numeracion
df['distrito'].iloc[1:] = df['distrito'].iloc[1:].str.slice(4).str.lower()
df
```

```python
# Cambiar tambien la mayuscula del valor Madrid y solo dejar en la fila de 'fuencarral-el pardo' solo 'fuencarral' y la fila 'moncloa-aravaca' solo 'moncloa'

df['distrito'] = df['distrito'].str.lower()
df['distrito'] = df['distrito'].str.replace('moncloa-aravaca', 'moncloa')
df['distrito'] = df['distrito'].str.replace('fuencarral-el pardo', 'fuencarral')
df['distrito'] = df['distrito'].str.replace('puente de vallecas', 'puente_vallecas')
df['distrito'] = df['distrito'].str.replace('ciudad lineal', 'ciudad_lineal')
df['distrito'] = df['distrito'].str.replace('villa de vallecas', 'villa_vallecas')
df['distrito'] = df['distrito'].str.replace('chamartín', 'chamartin')
df['distrito'] = df['distrito'].str.replace('tetuán', 'tetuan')
df['distrito'] = df['distrito'].str.replace('vicálvaro', 'vicalvaro')
df
```

```python
# cambiar los nombres de las columnas para que solo queden los años en numero y convertirlo en numerico
df = df.rename(columns={'2018 (p)': '2018', '2019 (a)': '2019'})
df['2018'] = pd.to_numeric(df['2018'], errors='coerce')
df['2019'] = pd.to_numeric(df['2019'], errors='coerce')
df
```

```python
# Pivotar la tabla para tener los distritos como columnas
pivot_df = df.set_index('distrito').T
pivot_df
```

```python
pivot_df = pivot_df.rename(columns={'distrito': 'año'})
pivot_df
```

```python
# Descargar el archivo 
archivo_excel = '/Users/Paco/IMMUNE/Modulo_Deep_Learning_Completo 2/Modulo_Deep_Learning_Completo/notebooks/renta_per_capita_limpia.xlsx'
pivot_df.to_excel(archivo_excel, index=True)
```

```python
# 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Obtener los datos de renta per cápita para un distrito específico (por ejemplo, 'madrid')
distrito = 'madrid'
distrito_data = df[df['distrito'] == distrito].iloc[:, 1:].values.squeeze()

# Crear una lista de años desde 2002 hasta 2019
years = range(2002, 2020)

# Crear una lista de años futuros para los que se harán las estimaciones
future_years = [2020, 2021, 2022]

# Ajustar un modelo de regresión lineal a los datos históricos
regression_model = LinearRegression()
X = [[year] for year in years]
y = distrito_data
regression_model.fit(X, y)

# Realizar las estimaciones para los años futuros utilizando el modelo de regresión lineal
future_X = [[year] for year in future_years]
predictions = regression_model.predict(future_X)

# Imprimir las estimaciones
print(f"Estimaciones de renta per cápita para el distrito {distrito}:")
for year, prediction in zip(future_years, predictions):
    print(f"Año {year}: {round(prediction, 2)}")

# Calcular el error cuadrático medio y el coeficiente de determinación
y_pred = regression_model.predict(X)
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

# Imprimir las métricas de evaluación
print(f"Raíz del error cuadrático medio: {rmse:.2f}")
print(f"Coeficiente de determinación: {r2:.2f}")

# desviacion estandar
# modelo
```

```python
desviacion_estandar = np.std(distrito_data)
print(f"Desviación estándar de renta per cápita para el distrito {distrito}: {desviacion_estandar:.2f}")
```

```python
# Crear el modelo para todos los distritos
# Crear una lista con todas las columnas
columnas = pivot_df.columns.tolist()

# Crear una lista para almacenar las filas del nuevo DataFrame
nuevas_filas = []

# Iterar sobre las columnas y aplicar el modelo
for columna in columnas:
    # Obtener los datos de la columna
    columna_data = pivot_df[columna].values.squeeze()
    
    # Ajustar un modelo de regresión lineal a los datos de la columna
    regression_model = LinearRegression()
    regression_model.fit([[year] for year in range(2002, 2020)], columna_data[:18])  # Utilizar solo los datos hasta 2019
    
    # Realizar las estimaciones para los años futuros utilizando el modelo de regresión lineal
    predictions = regression_model.predict([[year] for year in range(2020, 2023)])
    
    # Crear las nuevas filas para el DataFrame con las estimaciones
    nuevas_filas.append([columna] + predictions.tolist())

# Crear un nuevo DataFrame con las nuevas filas y las columnas correspondientes
nuevo_df = pd.DataFrame(nuevas_filas, columns=['Columna'] + [str(year) for year in range(2020, 2023)])

# Imprimir el nuevo DataFrame
nuevo_df
```

```python
# Pivoteo del DataFrame
pivot_df2 = nuevo_df.set_index('Columna').T

# Imprimir el DataFrame pivotado
pivot_df2
```

```python
# Añadir las nuevas filas al DataFrame existente pivot_df
df_addyears = pd.concat([pivot_df, pivot_df2])
df_addyears = df_addyears.astype(int)
df_addyears['Año'] = df_addyears.index
df_addyears = df_addyears.rename(columns={'index': 'Año'})
df_addyears = df_addyears.reset_index(drop=True)
df_addyears 
```

```python
# Convert it to a string and then to a datetime object with format 'DD/MM/YY'
df_addyears['Año'] = pd.to_datetime(df_addyears['Año'].astype(str), format='%Y').dt.strftime('%d/%m/%y')


# Print the updated DataFrame
df_addyears 
```

```python
df_addyears.rename(columns={'Año': 'Date'}, inplace=True)
df_addyears 
```

```python
# Utilizar el método melt para girar el DataFrame
df_renta_per_capita_girado = pd.DataFrame(df_addyears.melt(id_vars='Date', var_name='distrito', value_name='valor'))
df_renta_per_capita_girado
```
