## Paro

```python
import numpy as np
import pandas as pd

data_paro = '/Users/marioalvarogutierrez/Documents/Data Analytics/Capstone/Extracción/demograficos_paro.xlsx'

df_paro = pd.read_excel(data_paro)

df_paro
```

```python
df_paro = df_paro.rename(columns={'año': 'fecha'})
```

```python
df_paro.info()
```

Transformamos la columna año en tipo datetime mantiendo solo el año para facilitar el posterior analisis

```python
df_paro['fecha'] = pd.to_datetime(df_paro['fecha'], format='%Y')

df_paro.fecha.info()
```

```python
df_paro
```

Creamos una función tanto para este df como para el de calidad de vida para obtener la media entre los años 2019 y 2021 y añadirlo en la columna de 2020. Consideramos que el valor resultante de esta media es el que más se puede acercar al valor real.

```python
def añadir_valor_faltante(df):
    # Obtener los nombres de las columnas de los barrios
    columnas_barrios = df.columns[1:]

    # Iterar sobre los barrios
    for columna in columnas_barrios:
        # Obtener los valores de los años 2019 y 2021 en el barrio actual
        valor_2019 = df[df['fecha'] == '2019-01-01'][columna].values[0]
        valor_2021 = df[df['fecha'] == '2021-01-01'][columna].values[0]

        # Calcular la media entre los valores de 2019 y 2021
        media = (valor_2019 + valor_2021) / 2

        # Añadir el valor medio en el año 2020 del barrio actual
        df.loc[df['fecha'] == '2020-01-01', columna] = media

    return df

añadir_valor_faltante(df_paro)
```

```python
# Utilizar el método melt para girar el DataFrame
df_paro_girado = pd.DataFrame(df_paro.melt(id_vars='fecha', var_name='distrito', value_name='valor'))
```

```python
df_paro_girado.rename(columns={'valor': 'paro'}, inplace=True)

df_paro_girado
```

Multiplicamos por 10 los datos correspondientes a 2021 para igualarlos a los demás años que están medidos sobre 100 en vez de sobre 10

```python
df_cdv.loc[df_cdv['fecha'] == 2021, df_cdv.columns[1:]] *= 10

df_cdv
```

Transformamos la columna año en tipo datetime mantiendo solo el año para facilitar el posterior analisis

Predicción valor de 2022

Se intento predecir el valor correspondiente al año 2022 para cada distrito mediante una regresión lineal, pero las métricas no mostraron ser un modelo fiable para dicha predicción por lo que decidimos utilizar la media de los años anteriores en lugar de los obtenidos mediante la regresión. 

```python
# Eliminar la fila correspondiente al año 2022 que contiene valores NaN
df_cdv.dropna(subset=['fecha'], inplace=True)

# Calcular la media de cada columna para los años 2016 a 2021 y redondear a dos decimales
media_2016_2021 = df_cdv.loc[df_cdv['fecha'] <= '2021-01-01'].mean().round(2)

# Obtener el índice de la fila correspondiente al año 2022
indice_2022 = df_cdv[df_cdv['fecha'] == '2022-01-01'].index[0]

# Asignar los valores de la media de cada columna en la fila del año 2022
df_cdv.loc[indice_2022, df_cdv.columns[1:]] = media_2016_2021

df_cdv
```

```python
df_cdv
```

```python
# Utilizar el método melt para girar el DataFrame

df_cdv_girado = pd.DataFrame(df_cdv.melt(id_vars='fecha', var_name='distrito', value_name='valor'))
```

```python
df_cdv_girado.rename(columns={'valor': 'cdv'}, inplace=True)

df_cdv_girado
```

## Seguridad

```python
data_seguridad = '/Users/marioalvarogutierrez/Documents/Data Analytics/Capstone/Extracción/demograficos_seguridad.xlsx'

df_seguridad = pd.read_excel(data_seguridad)

df_seguridad
```

```python
df_seguridad = df_seguridad.rename(columns={'año': 'fecha'})
```

El primer año corresponde con la seguridad percibida en cada barrio durante el día y la segunda durante la noche, nos vamos a quedar con la media de ambas. Los años correspondientes con 2020 y 2021, que corresponden con la pandemia del COVID por lo que se asumen los datos de 2019.

```python
df_seguridad.info()
```

Multiplicamos por 10 los datos correspondientes a 2022 para igualarlos a los demás años que están medidos sobre 100 en vez de sobre 10

```python
df_seguridad.loc[df_seguridad['fecha'] == 2022, df_seguridad.columns[1:]] *= 10

df_seguridad
```

Realizamos la media de cada año, y es el unico dato con el que nos quedamos

```python
df_avg_seguridad = df_seguridad.groupby('fecha').mean()

df_avg_seguridad = df_avg_seguridad.reset_index()

df_avg_seguridad
```

```python
# Utilizamos el método melt para girar el DataFrame

df_avg_seguridad_girado = pd.DataFrame(df_avg_seguridad.melt(id_vars='fecha', var_name='distrito', value_name='valor'))
```

```python
df_avg_seguridad_girado.rename(columns={'valor': 'seguridad'}, inplace=True)

df_avg_seguridad_girado
```

Transformamos la columna año en tipo datetime mantiendo solo el año para facilitar el posterior analisis

```python
df_avg_seguridad_girado['fecha'] = pd.to_datetime(df_avg_seguridad_girado['fecha'], format='%Y')

df_avg_seguridad_girado.info()
```

```python
df_avg_seguridad_girado
```

Guardamos el archivo limpio

```python
# df_avg_seguridad_girado.to_csv('/Users/marioalvarogutierrez/Documents/Data Analytics/Capstone/Transformación/demograficos_avg_seguridad_long.csv', index=False)
```

### Creamos una tabla con los tres factores demograficos

```python
df_demograficos = pd.merge(df_paro_girado, df_cdv_girado, on=['fecha', 'distrito'])
df_demograficos = pd.merge(df_demograficos, df_avg_seguridad_girado, on=['fecha', 'distrito'])

df_demograficos
```

Transformamos la variable 'fecha' a dd/mm/aa y la renombramos como 'date' para igualarla a las demás tablas y facilitar su lectura, coexión y análisis. A su vez también creamos una columna year que recoja solo el año, con el mis o proposito.

```python
df_demograficos['year'] = df_demograficos['fecha'].dt.year
df_demograficos['date'] = pd.to_datetime(df_demograficos['fecha']).dt.strftime('%d/%m/%y')
df_demograficos = df_demograficos.drop(columns='fecha')

df_demograficos
```

Reducimos los decimales de todas la columnas de nuestroi df a 2

```python
df_demograficos = df_demograficos.round(2)

df_demograficos
```

## Apendice Zonas Verdes

Los datos desde 2020 hasta 2022 pertenecen al informe recogido en 2021

```python
data_zverde = '/Users/marioalvarogutierrez/Documents/Data Analytics/Capstone/Extracción/demograficos_zonas_verdes.xlsx'

df_zverde = pd.read_excel(data_zverde)

df_zverde
```

```python
df_zverde = df_zverde.rename(columns={'año': 'fecha'})
```

```python
df_zverde['fecha'] = pd.to_datetime(df_zverde['fecha'], format='%Y')

df_zverde.fecha.info()
```

```python
df_zverde
```

Añadimos el valor faltante de 2018 como la media de 2017 y 2018 para intentar acercarno lo máximo posible al dato real

```python
columnas_barrios = df_zverde.columns[1:]

# Iterar sobre los barrios
for columna in columnas_barrios:
    # Obtener los valores de los años 2017 y 2019 en el barrio actual
    valor_2017 = df_zverde[df_zverde['fecha'] == '2017-01-01'][columna].values[0]
    valor_2019 = df_zverde[df_zverde['fecha'] == '2019-01-01'][columna].values[0]

    # Calcular la media entre los valores de 2017 y 2019
    media = (valor_2017 + valor_2019) / 2

    # Añadir el valor medio en el año 2018 del barrio actual
    df_zverde.loc[df_zverde['fecha'] == '2018-01-01', columna] = media

# Verificar el resultado
df_zverde
```

```python
# Utilizamos el método melt para girar el DataFrame

df_zverde_girado = pd.DataFrame(df_zverde.melt(id_vars='fecha', var_name='distrito', value_name='valor'))
```

Cambiamos el nombre de valor a zverde

```python
df_zverde_girado.rename(columns={'valor': 'zverde'}, inplace=True)

df_zverde_girado
```

Igualamos las columnas correspondientes a las fechas para unirlo con el dataframe con todos los datos demograficos

```python
df_zverde_girado['fecha'] = pd.to_datetime(df_zverde_girado['fecha'], format='%Y')
df_zverde_girado['year'] = df_zverde_girado['fecha'].dt.year
df_zverde_girado['date'] = pd.to_datetime(df_zverde_girado['fecha']).dt.strftime('%d/%m/%y')
df_zverde_girado = df_zverde_girado.drop(columns='fecha')

df_zverde_girado
```

Lo unimos con la tabla master de los datos demograficos

```python
df_demograficos = pd.merge(df_demograficos, df_zverde_girado, on=['year', 'date', 'distrito'])

df_demograficos
```

Reorganizamos los datos 

```python
df_demograficos = df_demograficos[['distrito', 'paro', 'cdv', 'seguridad', 'zverde', 'year', 'date']]
df_demograficos = df_demograficos.round(2)

df_demograficos.head()
```
