### Limpieza y manipulación de tabla ventas

```python
#Importación de librerías
import pandas as pd
import numpy as np
```

```python
#Carga de datos
df_google = pd.read_csv("C:/Users/Adrian.Gallego/Desktop/google_04_23.csv")
```

```python
#Quitamos símbolos y espacios en las columnas 'var_mensual','var_trimest', 'var_anual'
df_google = df_google.rename(columns={'Mes': 'date'})
df_google = df_google.rename(columns={'alquiler madrid: (España)': 'alquiler_madrid'})
df_google = df_google.rename(columns={'Comprar Madrid: (España)': 'comprar_madrid'})
df_google = df_google.rename(columns={'alquiler Barcelona: (España)': 'alquiler_barcelona'})
df_google = df_google.rename(columns={'Comprar Barcelona: (España)': 'comprar_barcelona'})
df_google
```

```python
#Cambios de formato fecha: Cambiamos nombre de columna mes a date y creamos columna month y 

df_google["date"] = pd.to_datetime(df_google["date"])

# Crear la columna "mes" con el mes extraído de la columna "fecha"
df_google['moth'] = df_google['date'].dt.month

# Crear la columna "año" con el año extraído de la columna "fecha"
df_google['year'] = df_google['date'].dt.year
```

```python
#Exportamos la tabla limpiada
df_google.to_csv("C:/Users/Adrian.Gallego/Desktop/google_04_23.csv", index=False)
```
